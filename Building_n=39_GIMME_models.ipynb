```python
import cobra
from cobra.io import read_sbml_model
from mewpy.simulation import get_simulator
from mewpy.omics import ExpressionSet
from mewpy.omics import GIMME
import pandas as pd
from pandas import DataFrame
```

```python
%store -r expression #import CCLE 2018 dataframe
t_2018 = expression
t_2018
```

```python
model = read_sbml_model('Human-GEM-annotated.xml') #read Human1 model
```

```python
GIMME_CCLE_dataframe = pd.DataFrame(columns = ['Reaction ID'])
reaction_ids = []
for r in model.reactions:
    reaction_ids.append(r.id)
GIMME_CCLE_dataframe['Reaction ID'] = reaction_ids
GIMME_CCLE_dataframe = GIMME_CCLE_dataframe.set_index('Reaction ID')
GIMME_CCLE_dataframe
```

```python
#For the cell lines in gene expression dataset, iterate through and add fluxes to GIMME dataframe (above)
#Iterate through one media condition at a time and print as you go as can take a while
RPMI_cell_lines = ['JHOC5','KURAMOCHI','NIHOVCAR3','OC314','OVCAR8',
                   'OVISE','OVK18','OVKATE','OVMANA','OVSAHO','OVTOKO']

%store -r RPMI1640
envcond = RPMI1640

simul = get_simulator(model, envcond = envcond)

for c in RPMI_cell_lines:
    genes = expression.index.values
    cell_line = [c]
    exp = expression[cell_line].to_numpy()

    expr = ExpressionSet(genes, cell_line, exp)

    g = GIMME(simul, expr, parsimonious= True)

    cell_line_dataframe = g.dataframe
    GIMME_CCLE_dataframe[c] = (cell_line_dataframe.iloc[:,0]).to_list()
    print(c,'done')
    
GIMME_CCLE_dataframe
```

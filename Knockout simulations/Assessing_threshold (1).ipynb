```python
data = pd.read_csv(r'knockout_sim_results_with_stats_tab.csv')
data = data.set_index('genes')
data = data.iloc[:,7:]
data
```

```python
threshold = 0.95
proportions = []
for col in data.columns:
    proportion = (data[col] <= threshold).mean()
    print(f"{col}: {proportion:.4f}")
    proportions.append(proportion)
print('the average proportion of genes under this threshold is:',sum(proportions)/len(proportions))
```
